```python
from transformers import pipeline
```

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}
```

```python
generator_bert = pipeline(
    "text-generation",
    model=models["BERT"]
)

generator_bert("The future of Artificial Intelligence is")
```

```python
generator_roberta = pipeline(
    "text-generation",
    model=models["RoBERTa"]
)

generator_roberta("The future of Artificial Intelligence is")
```

```python
generator_bart = pipeline(
    "text-generation",
    model=models["BART"]
)

generator_bart("The future of Artificial Intelligence is", max_length=50)
```

MASKING

```python
fill_mask_bert = pipeline(
    "fill-mask",
    model=models["BERT"]
)

fill_mask_bert("The goal of Generative AI is to [MASK] new content.")
```

```python
fill_mask_roberta = pipeline(
    "fill-mask",
    model=models["RoBERTa"]
)

fill_mask_roberta("The goal of Generative AI is to <mask> new content.")
```

```python
fill_mask_bart = pipeline(
    "fill-mask",
    model=models["BART"]
)

fill_mask_bart("The goal of Generative AI is to <mask> new content.")
```

QUESTION ANSWERING

```python
qa_bert = pipeline(
    "question-answering",
    model=models["BERT"]
)

qa_bert(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

```python
qa_roberta = pipeline(
    "question-answering",
    model=models["RoBERTa"]
)

qa_roberta(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

```python
qa_bart = pipeline(
    "question-answering",
    model=models["BART"]
)

qa_bart(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```
