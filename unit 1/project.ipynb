```python
from transformers import pipeline

# Load Named Entity Recognition pipeline
ner_pipeline = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

resume_text = """
Anirudh Muraleedharan is a Computer Science student at Amrita Vishwa Vidyapeetham.
He previously interned at Infosys and is currently working with TCS on a web project.
"""

entities = ner_pipeline(resume_text)

for entity in entities:
    print(entity)
```

```python
person_name = None
organizations = []

for e in entities:
    if e['entity_group'] == 'PER':
        person_name = e['word']
    elif e['entity_group'] == 'ORG':
        organizations.append(e['word'])

print("Name:", person_name)
print("Organizations:", organizations)
```
